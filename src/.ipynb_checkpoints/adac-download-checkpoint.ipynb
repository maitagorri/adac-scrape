Ich glaube wir k√∂nnen es uns einfach machen--die URL-Struktur sieht wie folgt aus:

`https://www.adac.de/rund-ums-fahrzeug/autokatalog/marken-modelle/[MARKE]/[MODELL]/[UNTERMODELL]/[ID]`

Nur einige Modelle haben dann noch den Unterpunkt `/#kosten`

```python
## timestamp
# current date and time
now = datetime.now()
timestamp = now.strftime("%Y%m%d-%H%M")
```

```python
## Parameter
outpath='../out/'+timestamp+'_adac_data.csv'
```

## Liste von Modellen

```python
## Basisseite lesen
r = requests.get('https://www.adac.de/rund-ums-fahrzeug/autokatalog/marken-modelle/?filter=NONE')
base_soup = BeautifulSoup(r.text, 'html.parser')

# alle Marken finden
marken_urls = ['https://www.adac.de'+a.get('href') for a in base_soup.find_all('a', class_='sc-bdvvtL goIptw sc-gsDKAQ sc-faIbUi hIxhWw eunNEY')]
```

```python
alle_modelle = []
for marken_url in marken_urls:
    ## Marke
    marke = re.findall('/([^/]+)/[^/]*$', marken_url)[0]
    print(marke)
    
    r_marke = requests.get(marken_url)

    marke_soup = BeautifulSoup(r_marke.text, 'html.parser')

    # alle Baureihen finden
    baureihen_urls = ['https://www.adac.de'+a.get('href') for a in marke_soup.find_all('a', class_ = "sc-bdvvtL goIptw sc-gsDKAQ sc-gLEhor hIxhWw gwWXVZ")]


    for baureihe_url in baureihen_urls:
        baureihe = re.findall('/([^/]+)/[^/]*$', baureihe_url)[0]
        print('\t'+baureihe)
        ## Baureihe
        r_baureihe = requests.get(baureihe_url)
        baureihe_soup = BeautifulSoup(r_baureihe.text, 'html.parser')
        # alle Generationen finden
        generation_urls = ['https://www.adac.de'+a.get('href') for a in baureihe_soup.find_all('a', class_ = 'sc-bdvvtL goIptw sc-gsDKAQ sc-gLEhor hIxhWw gwWXVZ')]

        for generation_url in generation_urls:
            generation = re.findall('/([^/]+)/[^/]*$', generation_url)[0]
            print('\t\t'+generation)
            ## Generation
            r_generation = requests.get(generation_url)
            generation_soup = BeautifulSoup(r_generation.text, 'html.parser')
            # alle Modelle finden
            modelle_urls = ['https://www.adac.de'+a.get('href') for a in generation_soup.find_all(
                'a', 
                class_ = 'sc-bdvvtL goIptw sc-gsDKAQ sc-dqhCBx hIxhWw kxA-dmq')]
            
            alle_modelle = alle_modelle + modelle_urls

     
```

## Modellliste speichern, oder laden

```python
modelfile = '../out/'+datetime.now().strftime("%Y%m%d-%H%M")+'_modellist'
with open(modelfile, 'wb') as fp:
    pickle.dump(alle_modelle, fp)
```

```python
def downloadModelle(alle_modelle):
    print('Zu ladende Modelle: {}'.format(len(alle_modelle)))
    try: # der gesamte Prozess, wenn alles gut l√§uft
        i=0
        for modell_url in alle_modelle:
            if i==9:
                raise(IndexError)
            if i%100==0:
                print('... {}'.format(i))
            i += 1
            ## Modell
            r_modell = requests.get(modell_url + '#technische-daten')

            modell_soup = BeautifulSoup(r_modell.text, 'html.parser')

            technische_daten = modell_soup.find_all('div', 
                                                   class_ = 'sc-cxVPaa gxVuRJ sc-bnOPBZ jbcYSh sc-iKMXQg iMhTHO')
            tabellen = technische_daten[0].find_all('table')

            ## Daten zusammensetzen
            model_id = re.findall('/(\d+)/$',modell_url)[0]

            # Tabellen parsen
            df = pd.DataFrame()
            for t in tabellen:
                kategorien = [kategorie.text for kategorie in t.find_all('td', class_='sc-hDzdEj kyOEig')]
                werte = [wert.text for wert in t.find_all('td', class_='sc-hDzdEj cGvlQj')]
                df = pd.concat([df, pd.DataFrame({'Kategorie': kategorien, 'Wert':werte})])
            df['ADAC_id'] = model_id

            # Technische Daten in Datei schreiben
            df.to_csv(outpath, index=False, mode='a', header=not (os.path.exists(outpath)) )
    
    except IndexError: # wenn Abbruch: berichten, und mit dem Rest weitermachen
        # checken wie weit wir gekommen sind
        with open(outpath, 'r') as f:
            last_line = f.readlines()[-1]
            last_model = last_line.split(',')[-1].strip()
        print ("bekannter Fehler bei ID {}, Download wird neu aufgenommen".format(last_model))
        # bei welcher URL geht es weiter?
        for i,m in enumerate(alle_modelle):
            if(re.search(last_model,m)):
                break_index = i
        # weiter mit restlichen Modellen
        downloadModelle(alle_modelle[break_index:])
```
